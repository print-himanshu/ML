```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
data = np.random.normal(loc = 50, scale = 10, size = 10000 )
ax = plt.hist(data)
```

```python
bins_list  = [ -10, 20, 40, 50 ,60, 80, 110]
ax = plt.hist(data, bins = bins_list)
```

```python
ax = plt.hist(data, bins = 10)
```

```python
width = 3
n = math.ceil((data.max() - data.min())/width)
ax = plt.hist(data, bins = n)
```
