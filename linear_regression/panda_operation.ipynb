##  4. DataFrame.merge(right, how='inner' left_index=False, right_index=False, sort=False )
Merge DataFrame or named Series objects with a database-style join.<br>

The join is done on columns or indexes. If joining columns on columns, the DataFrame indexes will be ignored. Otherwise if joining indexes on indexes or indexes on a column or columns, the index will be passed on. When performing a cross merge, no column specifications to merge on are allowed.<br>

right: DataFrame or named Series<br>
&emsp;&emsp; - Object to merge with.

how: {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’<br>
&emsp;&emsp; - Type of merge to be performed.<br>

&emsp;&emsp;&emsp;&emsp; 1. left: use only keys from left frame, similar to a SQL left outer join; preserve key order.<br>

&emsp;&emsp;&emsp;&emsp; 2. right: use only keys from right frame, similar to a SQL right outer join; preserve key order.<br>

&emsp;&emsp;&emsp;&emsp; 3. outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.<br>

&emsp;&emsp;&emsp;&emsp; 4. inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.<br>

&emsp;&emsp;&emsp;&emsp; 5. cross: creates the cartesian product from both frames, preserves the order of the left keys.<br>


left_index: bool, default False<br>
&emsp;&emsp; - Use the index from the left DataFrame as the join key(s). If it is a MultiIndex, the number of keys in the other DataFrame (either the index or a number of columns) must match the number of levels.<br>

right_index : bool, default False<br>
&emsp;&emsp; - Use the index from the right DataFrame as the join key. Same caveats as left_index.<br>


## 5. DataFrame.sort_values(by, axis=0, ascending=True, inplace=False )

Sort by the values along either axis.<br>
 
by: str or list of str<br>
&emsp;&emsp; - Name or list of names to sort by.<br>

&emsp;&emsp;&emsp; 1. if axis is 0 or ‘index’ then by may contain index levels and/or column labels.<br>

&emsp;&emsp;&emsp; 2. if axis is 1 or ‘columns’ then by may contain column levels and/or index labels.<br>

axis: {0 or ‘index’, 1 or ‘columns’}, default 0<br>
&emsp;&emsp; - Axis to be sorted.<br>

ascending: bool or list of bool, default True<br>
&emsp;&emsp; - Sort ascending vs. descending. Specify list for multiple sort orders. If this is a list of bools, must match the length of the by.<br>

inplace: bool, default False<br>
&emsp;&emsp; - If True, perform operation in-place.<br>
